```python
import pickle as pkl
parkInfo = pkl.load(open('parkInfo.pkl'))
```

```python
categories = {}
area = {}

for park in parkInfo.iteritems():
    parkId = park[0]
    category = park[1]['Category']
    acres = park[1]['Acres']
    
    if categories.get(category) == None:
        categories[category] = 0
    categories[category]+=1
    
    if area.get(acres) == None:
        area[acres] = 0
    area[acres]+=1
```
